```python

import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download("vader_lexicon")
import pandas as pd
data = pd.read_csv("7282_1.csv")
# data = data.dropna()
print(data.head())
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data = data[['name','reviews.text']]
```

```python
data = data.dropna()
```

```python
sentiments = SentimentIntensityAnalyzer()
data["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in data["reviews.text"]]
data["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in data["reviews.text"]]
data["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in data["reviews.text"]]
data['Compound'] = [sentiments.polarity_scores(i)["compound"] for i in data["reviews.text"]]
data.head()
```

```python
score = data["Compound"].values
sentiment = []
for i in score:
    if i >= 0.05 :
        sentiment.append('Positive')
    elif i <= -0.05 :
        sentiment.append('Negative')
    else:
        sentiment.append('Neutral')
data["Sentiment"] = sentiment
data.head()
```

```python
# Group by hotel_name and sentiment, then count the occurrences of each sentiment per hotel
#This will give out positve , negative and neutral review for all hotels
sentiment_counts = data.groupby(['name', 'Sentiment']).size().unstack(fill_value=0)
```
